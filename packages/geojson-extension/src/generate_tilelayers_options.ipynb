```python
import xyzservices.providers as basemaps
import json

def define_tilelayer_options_dict():
    data = {}

    for [key, val] in basemaps.items():
        if ('url' in val):
            basemap_name = val.name
            if('max_zoom' in val):
                if('min_zoom' in val):
                    data[basemap_name]={'name': basemap_name, 'url':val.url, 'attribution': val.attribution, 'maxZoom': val.max_zoom, 'minZoom': val.min_zoom}
                else: 
                    data[basemap_name]={'name': basemap_name, 'url':val.url, 'attribution': val.attribution, 'maxZoom': val.max_zoom, 'minZoom': 0}
            else:
                data[basemap_name]={'name': basemap_name, 'url':val.url, 'attribution': val.attribution, 'maxZoom': 18, 'minZoom': 0}
        else:
            newdata = basemaps[key]

            for newval in newdata.values():
                basemap_name = newval.name
                if('max_zoom' in val):
                    if('min_zoom' in val):
                        data[basemap_name]={'name': basemap_name, 'url':newval.url, 'attribution': newval.attribution, 'maxZoom': newval.max_zoom, 'minZoom': newval.min_zoom}
                    else: 
                        data[basemap_name]={'name': basemap_name, 'url':newval.url, 'attribution': newval.attribution, 'maxZoom': newval.max_zoom, 'minZoom': 0}
                else:
                    data[basemap_name]={'name': basemap_name, 'url':newval.url, 'attribution': newval.attribution, 'maxZoom': 18, 'minZoom': 0}
                                       
    
    print(data)
    with open('tilelayers_options.json', 'w') as fp:
        json.dump(data, fp)               
```
